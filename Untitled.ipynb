```python
import numpy as np
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
# Load Data
df = pd.read_csv('NFLX.csv')
df.head()
```

```python
df.shape
```

```python
# Close Price
plt.figure(figsize=(16,8))
plt.title("Netflix")
plt.xlabel("Days")
plt.ylabel("Close Price($)")
plt.plot(df['Close'])
```

```python
# get close Price
df = df[['Close']]
df.head()
```

```python
# Var to predict 'x' days out into future shifted 'x' units/days up
f_days = 25
# target
df['Prediction'] = df[['Close']].shift(-f_days)
df.head()
```

```python
df.tail()
```

```python
# create feature data set (X) and convert to numpy array and remove last 'x' rows/days
X = np.array(df.drop(['Prediction'],1))[:-f_days]
print(X)
```

```python
# Create target dataset (y) convert to numpy array and get all target values except last 'x' rows/days
y = np.array(df['Prediction'])[:-f_days]
print(y)
```

```python
# split data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Create model
tree = DecisionTreeRegressor().fit(x_train,y_train)
lr = LinearRegression().fit(x_train,y_train)
```

```python
# Get last x rows of feature dataset
x_future = df.drop(['Prediction'], 1)[:-f_days]
x_future = x_future.tail(f_days)
x_future = np.array(x_future)
x_future
```

```python
tree_pred = tree.predict(x_future)
print(tree_pred)
print()
lr_pred = lr.predict(x_future)
print(lr_pred)
```

```python
# Visualize
pred = tree_pred
valid = df[X.shape[0]:]
valid['Predictions'] = pred
plt.figure(figsize=(16,6))
plt.xlabel('Days')
plt.ylabel('Close Price($)')
plt.plot(df['Close'])
plt.plot(valid[['Close', 'Prediction']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```

```python
# Visualize
pred = lr_pred
valid = df[X.shape[0]:]
valid['Predictions'] = pred
plt.figure(figsize=(16,6))
plt.xlabel('Days')
plt.ylabel('Close Price($)')
plt.plot(df['Close'])
plt.plot(valid[['Close', 'Prediction']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```

```python
lr.score(x_test,y_test)
```

```python
tree.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
